# Visualizations

```python
# Set Style
sns.set(style="whitegrid")
```

Survival Count by Gender(Count Plot)

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x='Sex',hue='Survived',palette='Set2')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived',labels=['No','Yes'])
plt.tight_layout()
plt.show()
```

Survival Count by Passenger Class(Count Plot)

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Pclass', hue='Survived', palette='Set1')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

HistoGram Plot

```python
plt.figure(figsize=(8,4))
sns.histplot(data=df,x='Age',bins=30,kde=True,hue='Survived',element="step",palette='Set1')
plt.title('Age Distribution by Survival')
plt.tight_layout()
plt.show()
```

Box Plot

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df,x='Survived',y='Fare',hue='Survived',palette='Set2')
plt.title('Fare Distribution by Survival')
plt.xticks([0, 1],['Not Survived','Survived'])
plt.tight_layout()
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='Age',y='Fare',hue='Survived',palette='Set1',alpha=0.7)
plt.title('Scatter Plot of Age vs Fare (Colored by Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived')
plt.tight_layout()
plt.show()
```

Heat Map

```python
plt.figure(figsize=(8, 6))
corr = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
